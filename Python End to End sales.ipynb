# 2. Performing Customers/sellers Segmentation

## a. Divide the customers into groups based on the revenue generated

```python
cust_order = pd.merge(customers , orders, how ='inner', on= 'customer_id' )
```

```python
cust_order_items = pd.merge(cust_order,order_items ,how = 'inner',on = 'order_id')
```

```python
cust_revenue = cust_order_items.groupby('customer_id')[['price']].sum()
cust_revenue.rename(columns ={'price':'total_spend_amt'} , inplace = True)
cust_revenue.head()
```

```python
cust_revenue['cust_rev_group'] = pd.cut(cust_revenue.total_spend_amt,
                               bins= np.arange(0 , 15000, 1000),
                               labels=['0-1000','1000-2000','2000-3000','3000-4000',
                                       '4000-5000','5000-6000','6000-7000','7000-8000','8000-9000',
                                      '9000-10000','10000-11000','11000-12000','12000-13000','13000-14000'])
```

```python
cust_revenue
```

## b. Divide the sellers into groups based on the revenue generated

```python
sellers_ord = pd.merge(order_items,sellers,on='seller_id', how = 'inner')
sellers_ord.head()
```

```python
sellers_rev = sellers_ord.groupby('seller_id')[['price']].sum()
sellers_rev.head()
```

```python
sellers_rev.rename( columns = {'price':'revenue'}, inplace= True)
```

```python
sellers_rev.max()
```

```python
sellers_rev['sellers_rev_group'] = pd.cut(sellers_rev.revenue,
                               bins= np.arange(0 , 300000, 50000),
                               labels=['0-50000','50000-100000','100000-150000','150000-200000',
                                       '200000-250000'])
```

```python
sellers_rev.head()
```

# 5. Customer satisfaction towards category & product

## a.Which categories (top 10) are maximum rated & minimum rated?

```python
orders_rating = pd.merge(order_review_ratings,order_items,how = 'inner',on ='order_id')
```

```python
prod_rating = pd.merge(orders_rating,products,how='inner',on= 'product_id')
prod_rating.head()
```

```python
## top 10 max rated category
prod_rating.groupby('product_category_name')[['review_score']].sum().sort_values(by='review_score',ascending = False).head(10)
```

```python
## top 10 minimum rated category
prod_rating.groupby('product_category_name')[['review_score']].sum().sort_values(by='review_score',ascending = True).head(10)
```

## b.Which products(top 10) are maximum rated & minimum rated?

```python
## top 10 max rated products
prod_rating.groupby('product_category_name')[['review_score']].sum().sort_values(by='review_score',ascending = False).head(10)
```

```python
## top 10 minimum rated products
prod_rating.groupby('product_category_name')[['review_score']].sum().sort_values(by='review_score',ascending = True).head(10)
```
